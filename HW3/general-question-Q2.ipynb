```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Q2 - General Question

```python
T = 7/12
N = 28
dt = 1/48 # T/N
S0 = 20
u = 1.04
d = 1/u
```

```python
nodes = np.zeros(N+1, dtype=float)
nodes[0] = S0

plt.scatter(0, S0, s=10)
for i in range(1, N+1):
    for j in range(i, 0, -1):
        nodes[j] = nodes[j-1] * u
    nodes[0] = nodes[0] * d
    x = [i] * (i+1)
    plt.scatter(x, nodes[:i+1], s=10)

plt.title("Q2-General Question")
plt.ylabel("S(t)")
plt.xlabel("steps")
plt.ylim(0,70)
plt.xlim(0,29)
plt.xticks([i for i in range(0, 29, 1)])
plt.plot()
plt.savefig("Q2-general-question.png")
```
